# Write row file

### Step1: dict_map_EntrezGeneID2ENSG [.biomart]

```python
# Ensembl Gene ID	EntrezGene ID	HGNC symbol	HGNC ID(s)	Gene type
# ENSG00000197468				unprocessed_pseudogene
# ENSG00000231049		OR52B5P	HGNC:15210	unprocessed_pseudogene
# ENSG00000276385				processed_pseudogene
# ENSG00000275151				processed_pseudogene
# ENSG00000228913	10537	UBD	HGNC:18795	protein_coding
# ENSG00000231948	100874343	HS1BP3-IT1	HGNC:41474	sense_intronic

dict_map_EntrezGeneID2ENSG = {}
non_unique_maps = {}

with open(file_biomart, 'r') as f:
    lines = f.readlines()
    for line_no, line in enumerate(lines):
        fields = line.split(",")
        ENSG = fields[0]
        EntrezGeneID = fields[1]
        if EntrezGeneID: # check that the field is non-empty
            if not EntrezGeneID in dict_map_EntrezGeneID2ENSG:
                dict_map_EntrezGeneID2ENSG[EntrezGeneID] = ENSG
            else:
                if not dict_map_EntrezGeneID2ENSG[EntrezGeneID] == ENSG: # Pascal suffers from being paranoid - and he knows it!
                    ### NOTE: WE ONLY KEEP THE FIRST OF THE MAPPINGS. The following mappings are NOT saved
                    #print "non-unique mapping detected"
                    non_unique_maps[EntrezGeneID] = ENSG
                    
print "number of non-unique mappings: {}".format(len(non_unique_maps))
print "mapped {} EntrezGeneID to ENSG".format(len(dict_map_EntrezGeneID2ENSG))
```

```python
dict_map_EntrezGeneID2ENSG
```

### Step2: dict_map_probeID2EntrezGeneID [.annot]

```python
set_n_fields = set()
dict_map_probeID2EntrezGeneID = {}

with open(file_annot, 'r') as f:
    lines = f.readlines()
    for line_no, line in enumerate(lines):
        fields = line.split("\t")
        if line.startswith("HEEBO"):
            probeID = fields[0]
            #gene_symbol = fields[2] # "HGNC"
            gene_symbol = fields[8] # GI
            if not probeID in dict_map_probeID2EntrezGeneID: # we have a newcomer! welcome to the dict!
                dict_map_probeID2EntrezGeneID[probeID] = gene_symbol
            else:
                if not (dict_map_probeID2EntrezGeneID[probeID] == gene_symbol): # lets make sure that the mapping are consistent. Warn...
                    print "The annot file does not agress about the mapping"
        #if "!platform_table_begin" in line:
        #    continue # skipping 
        set_n_fields.add(len(fields))
```

```python
dict_map_probeID2EntrezGeneID
```

### Step3: combine mapping: dict_map_probeID2ENSG

```python
dict_map_probeID2ENSG = {}
set_probe_mapped = set()
set_probe_unmapped = set()

for probeID in dict_map_probeID2EntrezGeneID:
    EntrezGeneID = dict_map_probeID2EntrezGeneID[probeID]
    print EntrezGeneID
    if EntrezGeneID in dict_map_EntrezGeneID2ENSG:
        ENSG = dict_map_EntrezGeneID2ENSG[EntrezGeneID]
        dict_map_probeID2ENSG[probeID] = ENSG
        set_probe_mapped.add(probeID)
    else:
        set_probe_unmapped.add(probeID)
        
print "Mapped probes: {}".format(len(set_probe_mapped))
print "Mapped unprobes: {}".format(len(set_probe_unmapped))
```
